```python
X=[1,2,3,4,5]
Y=[2,4,6,8,10]
plt.plot(X,Y)
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.plot(X,Y,color='red')
plt.plot(X,Y,linestyle='dashed')
plt.plot(X,Y,linewidth=2)
plt.plot(X,Y,marker='o')

plt.show()
```

```python
import matplotlib.pyplot as plt

X = [1, 2, 3, 4, 5, 6]
Y = [200, 300, 900, 500, 100, 700]

plt.plot(X, Y, color='green', linestyle='dashed', linewidth=2, marker='o')
plt.title("Line Plot")
plt.xlabel("Months")
plt.ylabel("Savings amount")
plt.show()
```

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D', 'E', 'F']
values = [10, 15, 9 ,7, 12, 9]

plt.bar(categories, values, color='skyblue')
plt.title("Basic Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
import matplotlib.pyplot as plt
ages=[20,22,22,23,25,27,27,28,30,31,32,34,35,35,43,45,47,49,51,53,55,57,59,60,65,57,69]
b=[20,30,40,50,60,70]
plt.hist(ages,bins=b,color='skyblue',edgecolor='black')
plt.title("Age Distribution of Customers")
plt.xlabel("Age Ranges")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

Languages = ['Python', 'Ruby', 'C++', 'Java']
Percentage = [40, 10, 20, 30]  # Use decimals for percentages
color = ["yellow", "pink", "lightgreen", "skyblue"]

plt.pie(Percentage, labels=Languages, colors=color, autopct='%1.1f%%',startangle=90)  # Display percentages
plt.legend()
plt.title("Programming Language Usage")
plt.show()
```

```python
import matplotlib.pyplot as plt

X = [1, 2, 3, 4, 5, 6]
Y = [200, 300, 900, 500, 100, 700]

plt.scatter(X, Y, color='green',s=100 , edgecolor='black')
plt.title("Scatter Plot Example")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```
